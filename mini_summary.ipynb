```python
import pdfplumber

pdf_file = '2024-PR-15.pdf'  # 실제 파일명 입력

try:
    with pdfplumber.open(pdf_file) as pdf:
        print(f"PDF 열기 성공!")
        print(f"총 페이지 수: {len(pdf.pages)}")
        
        full_text = ""
        for page_num, page in enumerate(pdf.pages):
            print(f"페이지 {page_num + 1} 처리 중...")
            text = page.extract_text()
            
            if text:
                print(f"  페이지 {page_num + 1}: {len(text)} 글자 추출")
                print(f"  첫 50글자: {text[:50]}")
                full_text += text
            else:
                print(f"  페이지 {page_num + 1}: 텍스트 없음")
        
        print(f"\n총 추출된 텍스트: {len(full_text)} 글자")
        
except FileNotFoundError:
    print("PDF 파일을 찾을 수 없습니다.")
except Exception as e:
    print(f"PDF 읽기 오류: {e}")
```

```python
def chunk_text(text, chunk_size=2000, overlap=100):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start = end - overlap
    return chunks

chunks = chunk_text(full_text)
print(f"청킹 완료!")
print(f"총 청크 개수: {len(chunks)}")
print(f"평균 청크 길이: {sum(len(chunk) for chunk in chunks) / len(chunks):.1f} 글자")

# 첫 번째 청크 확인
print(f"\n첫 번째 청크 (처음 300글자):")
print(chunks[0][:500])

# from langchain_text_splitters import RecursiveCharacterTextSplitter

# text_splitter = RecursiveCharacterTextSplitter(
#     chunk_size=1500,
#     chunk_overlap=200,
# )

# texts = text_splitter.split_documents(data)
# print(f"생성된 텍스트 청크 수: {len(texts)}")
# print(f"각 청크의 길이: {list(len(text.page_content) for text in texts)}")
```
